```python
#read data archive
df = pd.read_csv('/content/drive/MyDrive/Data Science/Trabalho/2022_Viagem.csv', encoding='latin-1', sep=';')

#show content
df
```

```python
#get info about panda analysis
df.info()
```

```python
df[df.isna().any(axis=1)]
```

```python
#Clean data
df['Justificativa Urgência Viagem'] = df['Justificativa Urgência Viagem'].fillna("Desconhecido")
df['CPF viajante'] = df['CPF viajante'].fillna("Desconhecido")
df['Cargo'] = df['Cargo'].fillna("Desconhecido")
df['Descrição Função'] = df['Descrição Função'].fillna("Desconhecido")

df[df.isna().any(axis=1)]
```

```python
df.info()
```

```python
df[df['Valor outros gastos'] == '?']
```

```python
df["Valor diárias"]=df["Valor diárias"].str.replace(',','.')
df["Valor passagens"]=df["Valor passagens"].str.replace(',','.')
df["Valor devolução"]=df["Valor devolução"].str.replace(',','.')
df["Valor outros gastos"]=df["Valor outros gastos"].str.replace(',','.')

df.info()
```

```python
df['Valor diárias'] = df['Valor diárias'].astype(float)
df['Valor passagens'] = df['Valor passagens'].astype(float)
df['Valor devolução'] = df['Valor devolução'].astype(float)
df['Valor outros gastos'] = df['Valor outros gastos'].astype(float)

df.info()
```

```python
df['Período - Data de início'] = pd.to_datetime(df['Período - Data de início'])
df['Período - Data de fim'] = pd.to_datetime(df['Período - Data de fim'])

df.info()
```

```python
print('Media diária', df['Valor diárias'].mean())

df.set_index('Período - Data de início')['Valor diárias'].plot()
```
